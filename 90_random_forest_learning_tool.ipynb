# **Random Forest Learning Tool**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1,random_state=14)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
# combined feature and row sampling to 0.5
combined_sample = df.sample(frac=0.5, axis=1, random_state=14).sample(frac=0.5, random_state=14)
```

```python
df1 = combined_sample
df2 = combined_sample.copy()
df3 = combined_sample.copy()
```

```python
df1.columns
```

```python
df3.columns
```

```python
df3.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
x1 = df1.iloc[:, 1:3]
y1 = df1['target']

x2 = df2.iloc[:, 1:3]
y2 = df2['target']

x3 = df3.iloc[:, 1:3]
y3 = df3['target']
```

```python
clf1.fit(x1, y1)
clf2.fit(x2, y2)
clf3.fit(x3, y3)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
